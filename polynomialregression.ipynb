```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
X = 11*np.random.rand(200, 1) - 7
y = 0.8*X**2 + 0.9*X + 2 + np.random.rand(200, 1)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
```

```python
model1 = LinearRegression()
model2 = SGDRegressor()
```

```python
model1.fit(X, y)
```

```python
model2.fit(X, y)
```

```python
print(r2_score(y, model1.predict(X)))
print(r2_score(y, model2.predict(X)))
```

```python
plt.scatter(X, y)
plt.plot(X, model1.predict(X), color = 'red')
plt.plot(X, model2.predict(X), color = 'green')
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2)
```

```python
X_train_trans = poly.fit_transform(X)
```

```python
model1.fit(X_train_trans, y)
```

```python
plt.scatter(X, y)
plt.plot(X, model1.predict(X_train_trans), color = 'red')
```

```python
print(r2_score(y, model1.predict(X_train_trans)))
```
